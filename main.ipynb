Opis zbioru danych \
Zbiór danych pochodzi z General Social Survey i obrazuje on nastroje społeczne w Stanach Zjednoczonych w temacie aborcji na przestrzeni lat 1972-2018. General Social Survey jest badaniem socjologicznym stworzonym w 1972 roku przez National Opinion Research Center na Uniwersytecie w Chicago i finansowanym przez National Science Foundation. GSS zbiera informacje co dwa lata i prowadzi historyczny zapis obaw, doświadczeń, postaw i praktyk mieszkańców Stanów Zjednoczonych. 
Zbiór danych zawiera informacje dotyczące opinii na temat aborcji motywowanej poszczególnymi sytuacjami, wraz z danymi dotyczącymi ankietowanych takimi jak ich wiek, płeć, rasa, poglądy polityczne, religijność, czy edukacja. Ponadto dane zawierają informacje na temat roku przeprowadzenia badania.





```r
data <- read.csv("gss_abortion.csv")
data$age <- as.numeric(data$age) #Zmiana typu danych kolumn age i educ na numeric
data$educ <- as.numeric(data$educ)
data$relactiv <- as.numeric(data$relactiv)
data <- na.omit(data) #Usuniemy wszystkie niekompletne wiersze, by brać pod uwagę jedynie kompletne badania
View(data)
```

```r
#W tym momencie mamy poglądową wiedzę na temat zbioru danych. Możemy więc przejść do obliczania prostych elementów statystyki opisowej
#Warto byłoby wiedzieć jak wygląda rozkład ilości badań w zależności od roku przeprowadzenia


# Przygotowanie danych
year_data <- data$year

# Stworzenie szeregu rozdzielczego
hist(year_data, breaks = "FD", main = "Szereg rozdzielczy - Rok", xlab = "Rok")


#Wykres pokazuje, że kompletne badania pochodzą jedynie z lat 2006-2018
```

```r
#Widzimy, że większość badań przeprowadzana była w XXI wieku. Użyjmy teraz funkcji summary, by dowiedzieć się więcej o danych.
summary(data)
#Z summary wynika, że średnia wieku ankietowanych to blisko 47 lat, a mediana 45. Warto mieć to na uwadze, gdyż wynika z tego że próba składa się z osób raczej w średnim wieku.
#Średnia i mediana lat spędzonych na edukacji jest równa 12. Biorąc pod uwagę, że obowiązek szkolny w USA trwa 12 lat, to możemy stwierdzić że większość pytanych osób ma średnie wykształcenie.
#Są to jednak tylko początkowe spostrzeżenia. W rzeczywistości rokzłady mogą wyglądać inaczej. Sprawdźmy jak dokładnie wyglądają rozkłady dla wieku i edukacji.
```

```r
age_count <- table(data$age)

barplot(age_count, xlab = "Wiek", ylab = "Ilość", main = "Rozkład wieku badanych")

#Rokzład wieku jest nieco inny niż założyliśmy wyżej. W rzeczywistości większość ankietowanych stanowią osoby młodsze. Cechuje się on lekką prawoskośnością.
```

```r
edu_count <- table(data$educ)

barplot(edu_count, xlab = "Wiek", ylab = "Ilość", main = "Rozkład wieku badanych")

#Tym razem byliśmy bliżej prawdy. Przytłaczająca większość badanych kształciła się dokładnie 12 lat. Niemniej istnieją również badani, którzy kontynuowali swoją edukację. Rozkład ten z kolei jest wyraźle lewoskośny.
```

```r
#Teraz sprawdźmy jak w ogóle wygląda rozkład jeżeli chodzi o opinie dotyczące prawa do aborcji.

yes_no_counts <- table(ifelse(data$abany == "1", "Tak", "Nie"))
barplot(yes_no_counts, ylab = "Ilość", main = "Rozkład opinii na temat prawa do aborcji")

#Większość badanych jest przeciwna prawa do aborcji.
```

```r
#Sprawdźmy teraz jak wyglądają zależności między opinią, a latami edukacji, wiekiem, oraz religijnością.
#Posłużymy się do tego regresją liniową
age_grouped <- aggregate(abany ~ age, data = data, FUN = mean)

plot(age_grouped$age, age_grouped$abany, 
     type = "p", col = "black", pch = 16, 
     xlab = "Wiek", ylab = "Proporcjonalne popieranie aborcji ",
     main = "Związek między wiekiem a poparciem aborcji")

model_age <- lm(abany ~ age, data = age_grouped)

abline(model_age, col = "red")

#Wyświetlmy również podsumowania modeli.
summary(model_age)
```

```r
#Zależność opinii od edukacji
education_grouped <- aggregate(abany ~ educ, data = data, FUN = mean)

plot(education_grouped$educ, education_grouped$abany, 
     type = "p", col = "black", pch = 16, 
     xlab = "Lata edukacji", ylab = "Proporcjonalne popieranie aborcji ",
     main = "Związek między edukacją a poparciem aborcji")

model_education <- lm(abany ~ educ, data = education_grouped)

abline(model_education, col = "blue")

summary(model_education)
```

```r
#Zależność opinii od religijności
relactiv_grouped <- aggregate(abany ~ relactiv, data = data, FUN = mean)

 plot(relactiv_grouped$relactiv, relactiv_grouped$abany, 
     type = "p", col = "black", pch = 16, 
     xlab = "Religijność w skali 1-11", ylab = "Proporcjonalne popieranie aborcji ",
     main = "Związek między religijnością a poparciem aborcji")    

model_relactiv <- lm(abany ~ relactiv, data = relactiv_grouped)

abline(model_relactiv, col = "green")

summary(model_relactiv)
```

Korzystając z podsumowania wszystkich modeli dochodzimy do wniosku, że w każdym z powyższych przypadków wyniki są istotne. Związki pierwszych dwóch modeli, tj. związki między wiekiem i edukacją a religijnością cechują się wartością p-value bliską zeru. W przypadku związku religijności a poparciem aborcji wynik ten jest nieco większy, lecz dalej wskazuje na związek między danymi.

```r
#Biorąc pod uwagę powyższe wartości należałoby jeszcze sprawdzić jaki wpływ ma postęp czasu na opinie społeczne na temat aborcji. 
#H0: Z biegiem lat lość osób opowiadających się za prawem do aborcji się nie zmienia
#H1: Z biegiem lat lość osób opowiadających się za prawem do aborcji się zmienia

#Mamy do czynienia z danymi nieparametrycznymi, które nie mogą być uporządkowane. Posłużymy się więc testem chi-kwadrat
alfa <- 0.05 #przyjmijmy alfa na poziomie 0.05
RokDecyzja <- table(data$year,data$abany) #Tworzymy tablicę wielodzielczą dla atrybutów rok i decyzji abany (1 oznacza poparcie)
RokDecyzja
plot(RokDecyzja) #możemy wyświetlić wykres tej tabeli przy pomocy funkji plot
```

```r
sum_rows <- margin.table(RokDecyzja, margin = 1) #Obliczmy sumę obserwacji i liczebności brzegowe
sum_cols <- margin.table(RokDecyzja, margin = 2)
n <- sum(RokDecyzja)

sum_rows
sum_cols
n

#Obliczmy stopnie swobody
df <- (length(sum_rows) - 1) * (length(sum_cols) - 1)
print(paste("Liczba stopni swobody: " ,df))
#I tabelę wartości oczekiwanych
E <- outer(sum_rows, sum_cols)/n
E
```

```r
#Obliczmy wartość statystyki testowej
chi2  <- sum((RokDecyzja - E)^2/E)
print(paste("Wartość statystyki chi-kwadrat: " ,chi2))
#W celu oceny wyniku wyliczmy wartość krytyczną
kryt <- qchisq(alfa, df, lower.tail = F)
print(paste("Wartość krytyczna:  " ,kryt))

#Na podstawie obliczonych wartości możemy odrzucić hipotezę H0 na rzecz hipotezy H1. Wynika z tego, że z biegiem lat lość osób opowiadających się za prawem do aborcji zmienia się.
```
